## Task 12. Show null values

```python
df.isnull().sum()
```

We can see that the output is truncated. Therefore, I want to see only the rows which has null values, and without truncation. So, I will convert the series into dictionary. 

```python
null_values = df.isnull().sum()
null_values = null_values[null_values > 0]
null_values = null_values.to_frame()
```

```python
null_values.to_dict()
```

In this way, we can observe all rows as **dictionary keys**, and the number of null values in the corresponding rows as **dictionary values**
