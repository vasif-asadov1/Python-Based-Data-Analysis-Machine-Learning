Task 5.  Reverse the following list using for loop: 
list1 = [10, 20, 30, 40, 50] 

```python
list1 = [10,20,30,40,50]
copy_list = list1.copy()
l = len(list1)
for idx, i in enumerate(list1): 
    copy_list[l - idx - 1] = i 
list1 = copy_list
print(list1)



```

```python
# easier way 
list1 = [10,20,30,40,50]
list2 = []
for i in range(len(list1)-1,-1, -1):
    list2.append(list1[i])
list1 = list2 
del list2 
list1 
```

```python
# with while loop: 
list1 = [10,20,30,40,50]
i = len(list1)-1
list2 = []
while i>=0: 
    list2.append(list1[i])
    i-=1 
list1=list2 
del list2 
list1 
```

Task 22. This time using filter() and list() functions filter all the positive integers in the string. 

```python
s = "abc -5 8 -3 2 x9 -1 4 -0 7 10 -99"


def check_positivity(idx):  
    flag = False 
    if s[idx].isdigit() and (idx ==0 or s[idx - 1] != "-"): 
        flag = True 
    return flag 


# here we want to filter out the indices of positive digits
# positive digits must satisfy two conditions: 
# 1. must be number (check with isdigit())
# 2. must not have minus sign before it
# is_positive() function checks these conditions. 


def is_positive(char):  # function will receive the selected character inside the string
    idx = s.index(char) # it will find the index of this character
    flag = False    
    if char.isdigit() and (idx == 0  or s[idx-1] != "-"): # this checks whether that character is digit
        flag = True     # then checks whether it is first character or it does not have "-" sign before it. 
    return flag         # based on condition checking it will return flag (True or False)

result = list(filter(is_positive, s)) # filter function will eliminate characters inside the string based on
# the True/False results of is_positive function.

result 
```
