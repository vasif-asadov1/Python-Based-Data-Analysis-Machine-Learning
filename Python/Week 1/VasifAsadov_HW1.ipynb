Task 2 opt. Write a Python function to check number is positive, negative or zero.

```python
def check_sign(num): 
    sign = ""
    if num>0:
        sign = "positive"
    elif num==0:
        sign = "zero"
    else: 
        sign = "negative"
    return sign 
```
