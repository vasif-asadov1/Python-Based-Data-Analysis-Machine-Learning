<font color = "Blue">Task 8.  Capitalize the first letter, lowercase, uppercase, swapcase, title-case of all the elements of a given array. Use </font> **np.char.capitalize()**, **np.char.lower()**, **np.char.upper()**, 
**np.char.swapcase()**, **np.char.title()**.

````python
arr = np.array([["Baku", "Ismayilli", "shamakhi"], 
               ["guba","Gabala","Gobustan"], 
               ["Khocali", "aghdam", "Khocavend"]])

capitalized = np.char.capitalize(arr) 
lowercase = np.char.lower(arr)
uppercase = np.char.upper(arr) 
swapcase = np.char.swapcase(arr) 
titlecase = np.char.title(arr) 
````

````python
print("Capitalized matrix:\n ",capitalized,"\n\n")
print("Lowercase matrix:\n ",lowercase, "\n\n")
print("Uppercase matrix:\n ",uppercase,"\n\n")
print("Swapcase matrix:\n ",swapcase,"\n\n")
print("Titlecase matrix:\n ",titlecase,"\n\n")
````

Data columns:

<font color = "Blue">Task 1. Show relationship between highway and city miles per gallon. 

a. Use scatter in matplotlib.  
b. Interpret how are variables correlated according to scatter.  

</font>

````python
plt.figure(figsize=(8,4))
plt.scatter(cars["hwy"], cars["cty"])
plt.title("Highway mpg vs city mpg")
plt.xlabel("Highway mpg")
plt.ylabel("City mpg")
plt.show()

````

We can see the strong relationship between highway mpg and city mpg.
