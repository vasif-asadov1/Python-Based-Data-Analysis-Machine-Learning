<font color = "blue">Task 1. Which matches a word containing vowels in it (a,ı,o,u,e,ə,i,ö,ü). </font>

```python
string = 'sdkmfsdf sabri sdlf,sl_ climanjaro tableau waikiki trambolin ol$3344 sənətkar '
string 
```

```python
re.findall(r"\w*[aioueəiöü]\w*", string)
```

<font color = "blue">Task 2. Which matches a word containing vowels not at the start or end of the word 
(a,ı,o,u,e,ə,i,ö,ü).  </font>

```python

text = "ayati vakashivazu bakabayashi Əli Əhməd Heydər Hakim Rəsul Padişah Qəbələ İsmayıllı"

re.findall(r"\b[^ aıoueəiöüAIOUEƏİÖÜ]\S*[^ aıoueəiöüAIOUEƏİÖÜ]\b",text) 
```

<font color = "Blue">Task 3. Checks for a number at the end of a given string </font>

```python
text = "Sabah student tell7 seventeen please Area51 Baku RegEx Difficul4"

re.findall(r"\b\w*\d\b", text)
```

<font color = "Blue"> Task 4. Which matches a word at the end of string, with optional question mark and

```python
text = "ssdf! sdkfmsdmf!? sdkmfkdsf weroweor .s,sdf, sd 84 84541 4sdfds4 897sdfldm sldfl!! skdfm llsd558?"

re.findall(r"\b\w+[\?!]+", text)
```

<font color = "Blue"> Task 7.  Removes all whitespaces from a string.  </font>

<font color = "blue">Task 8. Finds all words which have more than 4 characters in a string. </font>

```python
text = "High low epic movie stand Go Trip rental science analytics knowledge math keep simulator Python"

re.findall(r"[A-z]{4,}",  text)
```

<font color = "Blue">Task 9. Inserts spaces between words starting with capital letters. </font> <br>
Original string: PythonProgrammingLanguage <br>
Expected output: Python Programming Language 

```python
text =  "PythonProgrammingLanguage"

re.sub(r"([a-z])([A-Z])", r"\1 \2", text)
# \1 captures the first group - the lowercase letter
# \2 captures the second group - the uppercase letter
# \1 \2 adds space between them 
```
